## 1. INTRODUCTION

It is said that the history of music is as old as humanity itself. By observing various attributes of music, including acoustics, danceability, energy, and liveliness, we will attempt to determine which, if any, specific feature(s) result in a hit song. To accomplish this, we seek to compare the results of several supervised regression machine learning algorithms, and assess the results of the best-fitting algorithm to our model. The insights from this project can not only help artists produce hit songs – which in itself would be revolutionary in how music is conceived and produced in the future – but also provide them with perceptive information about what qualities specifically bring about the emotions they wish to evoke through their art.

Our dataset has brought forth a multitude of questions. Of course, the most general and all-encompassing: which features, if any, can be used to determine song popularity? ### 3.1. Data Wrangling

This cell obtains the .csv file, containing the data, from a public Google Drive link, and then reads the .csv file into a pandas DataFrame.

```python
import pandas as pd
import requests
from io import StringIO

orig_url='https://drive.google.com/file/d/1l8MCvSMTtNDzlbPZ387OflXp1A21xH1U/view?usp=sharing'

file_id = orig_url.split('/')[-2]
dwn_url='https://drive.google.com/uc?export=download&id=' + file_id
url = requests.get(dwn_url).text
csv_raw = StringIO(url)
data = pd.read_csv(csv_raw)
```

This cell performs simple data cleaning, as well as data wrangling. It reads our dataset into a dataframe, drops insignifcant columns, replaces all null values with 0, and splits the dataframe into a dataframe for feature variables and a series for the target variable. This is done to eventually split the relevant data into training and testing sets.

```python
#dropping the columns that are irrelevant or cannot have an effect on our data such as the artist name 
data = data.drop(columns = ['id', 'release_date', 'name', 'year', 'artists'], axis = 1)

#dropping any rows with null values
data.dropna(axis = 0, inplace = True)

#splitting the dataframe into a data frame for the features and the target variable
features = data.drop('popularity', axis = 1)
target = data['popularity']
```

This cell first splits the data into training and testing sets, both for the feature variables and target variables. It then further splits the training data into training and validation sets for both the feature and target variables.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, target, random_state = 3000)
#PRINT OUTPUT OF X-TRAIN, X-TEST, Y...

X_train, X_validation, y_train, y_validation = train_test_split(X_train, y_train, test_size = 0.2, random_state = 3000)
```

This cell preprocesses the feature variables in the training, testing, and validation sets by transforming the feature variables using the StandardScaler. The StandardScaler was used to create a more normal distribution of the values. 

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train) 
X_test_scaled = scaler.transform(X_test) 
X_validation_scaled = scaler.transform(X_validation)
```

Using recursive feature elimination, this cell accomplishes feature selection in order to select the most significant features.The selected features in this case are acousticness, danceability, duration_ms, energy, instrumentalness, key, liveness, loudness, speechiness, tempo and valence.  

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.feature_selection import RFE
dtr = DecisionTreeRegressor()
rfe = RFE(estimator=dtr, n_features_to_select = 11)
rfe.fit(X=X_train_scaled, y=y_train)
X_train_selected = rfe.transform(X=X_train_scaled)
X_test_selected = rfe.transform(X=X_test_scaled)
X_validation_selected = rfe.transform(X=X_validation_scaled)
print("Selected features after RFE: ")
for f in range(len(features.columns)):
    if rfe.get_support()[f] :
        print("\t", features.columns[f])
```

### 3.3. Model Construction

Hypothesis Testing:

Hypothesis 1: Energy has a positive correlation with popularity.    
IV: energy         
DV: popularity              
Null Hypothesis: Energy and population have no correlation.         

Hypothesis 2: Danceability has a positive correlation with popularity.          
IV: danceability                  
DV: popularity                   
Null Hypothesis: Danceability and popularity have no correlation. 

The following cell uses the KBinsDiscretizer to bin the popularity, danceability, and energy columns. This process is most effective for linear models and leads to better results.

```python
from scipy import stats
from scipy.stats import chisquare
from sklearn.preprocessing import KBinsDiscretizer

# Binning the popularity column to be able to create a 'categorical' variable to use for the chi-square test. 
discretizer = KBinsDiscretizer(n_bins = 3, encode = 'ordinal', strategy = 'uniform')
discret_pop = discretizer.fit_transform(data['popularity'].values.reshape(-1,1))

data_discret_pop = data.copy()
data_discret_pop['popularity'] = discret_pop

# Binning the danceability column to be to create another 'categorical' variable for the chi-square test. 
discretizer = KBinsDiscretizer(n_bins = 10, encode = 'ordinal', strategy = 'uniform')
discret_dance = discretizer.fit_transform(data['danceability'].values.reshape(-1,1))
data_discret_pop['danceability'] = discret_dance


# Binning the energy column to be to create another 'categorical' variable for the chi-square test. 
discretizer = KBinsDiscretizer(n_bins = 10, encode = 'ordinal', strategy = 'uniform')
discret_energy = discretizer.fit_transform(data['energy'].values.reshape(-1,1))
data_discret_pop['energy'] = discret_energy
data_discret_pop
```

Running the Chi-Square Tests:

```python
# Creating the crosstab data frame to use to calculate chi-square stats for danceability and popularity. 
cross_tab_dance = pd.crosstab(data_discret_pop['danceability'], data_discret_pop['popularity'], margins = False)
cross_tab_dance
```

```python
# Calculating chi-square stats for danceability and popularity (Hypothesis 1)
from scipy.stats import chi2_contingency
stat, p, dof, expected = chi2_contingency(cross_tab_dance)
print('Statistic: ', stat)
print('p-value: ', p)
print('Degrees of Freedom: ', dof)
```

```python
# Creating the crosstab data frame to use to calculate chi-square stats for energy and popularity. 
cross_tab_energy = pd.crosstab(data_discret_pop['energy'], data_discret_pop['popularity'], margins = False)
cross_tab_energy
```

```python
# Calculating chi-square stats for energy and popularity (Hypothesis 2)
stat2, p2, dof2, expected2 = chi2_contingency(cross_tab_energy)
print('Statistic: ', stat2)
print('p-value: ', p2)
print('Degrees of Freedom: ', dof2)
```

The next three cells are using GridSearch to find the optimal paramaters for the K-neighbors regression model, the Ridge model and the Lasso model in that order. For the K-neighbors regression model, the distance metric and optimal number of neighbors are being tested. The best metric is determined to be manhattan and the optiminal number of neighbors is determined to be 50. 

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
metrics = ["euclidean", "manhattan"]
n_vals = [10,50,100]
param_grid = {'metric': metrics, 'n_neighbors': n_vals}
from sklearn.model_selection import GridSearchCV
grid_search = GridSearchCV(KNeighborsRegressor(n_jobs = -1), param_grid, cv=5)
grid_search.fit(X_train_selected, y_train)
print("Best parameters: ", grid_search.best_params_)
```

```python
# The grid search is optimizing the alpha values for the Ridge regression model. The best value for alpha for this model is
# determined to be 10. 
```

```python
alpha_vals = [1, 0.5, 5, 10, 0.1, 25, 60, 0.001, 0.0001]
param_grid2 = {'alpha': alpha_vals}
grid_search2 = GridSearchCV(Ridge(), param_grid2, cv=5)
grid_search2.fit(X_train_selected, y_train)
print("Best parameters for Ridge Model: ", grid_search2.best_params_)
```

```python
# The grid search is optimizing the alpha values for the lasso regression model. The best value for alpha for this model is 
# determined to be 0.0001. 
```

```python
grid_search3 = GridSearchCV(Lasso(), param_grid2, cv=5)
grid_search3.fit(X_train_selected, y_train)
print("Best parameters for Lasso Model: ", grid_search3.best_params_)
```

The next cell is training all of the ML models with the selected features and the target variable of popularity. 

```python
estimators_not_optimized = {'Linear Regression: ' : LinearRegression(),  
              'Ridge:' : Ridge(),
              'Lasso:' : Lasso(),
              'k-nearest Neighbor:' : KNeighborsRegressor(n_jobs = -1)}
for key, est in estimators_not_optimized.items():
    est.fit(X = X_train_selected, y = y_train)
```

### 3.4. Model Evaluation

```python
# The next cell is using the validation set to test the not - optimized models for evidence of over or underfitting. 
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
for key, est in estimators_not_optimized.items():
    print(key, "\n")
    print('\t' + 'R-squared value for training set: ', r2_score(y_train, est.predict(X_train_selected)), "\n")
    print('\t' + 'R-squared value for validation set: ', r2_score(y_validation, est.predict(X_validation_selected)), "\n")
    print('\t' + 'RMSE for training set: ', mean_squared_error(y_train, est.predict(X_train_selected)), "\n")
    print('\t' + 'RMSE value for validation set: ', mean_squared_error(y_validation, est.predict(X_validation_selected)), "\n")
```

The ML model with the best performance is the k-nearest Neighbors regression model as it has the both the highest r-squared value for both the training the testing set. Also, the RMSE values for the k-nearest Neighbors regression model are significantly lower than the RMSE values for all of the other models which also indicates that this model had the best performance. However, this does show evidence of overfitting as the r-squared for the training set is significantly higher than the R-squared value for the testing set meaning. This is also shown with the RMSE values as the RMSE for the training set is significantly lower than the RMSE value for the validation set. Because a lower RMSE value indicates a better fit, this is also evidence of overfitting. 
All of the other models show slight overfitting as the r-squared for the training sets are slightly larger than the r-squared values for the testing sets. The RMSE values also show this as they are slightly lower for the training sets. 

 

Hypothesis Testing: Both tests have a p value of 0.00 which is below 0.05 which means that they are both statistically significant and both null hypotheses should be rejected. This means that there is a correlation between energy and popularity and there is a correlation between the danceability and popularity. 

### 3.5. Model Optimization

```python
# The next cell is fitting the optimized models to the training data. 
```

```python
estimators = {'Linear Regression: ' : LinearRegression(),  
              'Ridge:' : Ridge(alpha = 25),
              'Lasso:' : Lasso(alpha = 0.0001),
              'k-nearest Neighbor:' : KNeighborsRegressor(metric = 'manhattan', n_neighbors = 50, n_jobs = -1)}
for key, est in estimators.items():
    est.fit(X = X_train_selected, y = y_train)
```

The ideal parameters or combination of parameters for each ML model were determined in 3.4(Model Construction) through GridSearch. These parameters were then inputted into this new list of ML models, creating hyper-parameter - tuned models. For the Ridge and Lasso models, the alpha levels were tuned as they were both slightly overfitted and changing the alpha level could fix this, so both of them were tested with a range of alpha values from 0.1 to 60 using GridSearch. The Ridge model ran the best with an alpha level of 25 and the Lasso model ran the best with an alpha level of 0.0001.
The k-nearest Neighbors model shown to be over-fitted in the last section so the the hyperparamaters that were tuned for the k-nearest Neighbors model were the n_neighbors and the metrics. Using GridSearch, with the n_neighbors being tested being 10 and 50, and the metrics being tested being euclidean and manhattan, the results were that the model performed the best with n_neighbors = 50, and metric = 'manhattan'. The n_jobs = -1 is too speed up the run time of this algorithm. 

### 3.6. Model Testing

```python
# The next cell is using the testing set to test the optimized and fitted and optimized models. They show improvement from 
# the unoptimized models. 
```

```python
from sklearn.metrics import r2_score
for key, est in estimators.items():
    print(key, "\n")
    print('\t' + 'R-squared value for training set: ', r2_score(y_train, est.predict(X_train_selected)), "\n")
    print('\t' + 'R-squared value for testing set: ', r2_score(y_test, est.predict(X_test_selected)), "\n")
    print('\t' + 'RMSE for training set: ', mean_squared_error(y_train, est.predict(X_train_selected)), "\n")
    print('\t' + 'RMSE value for testing set: ', mean_squared_error(y_test, est.predict(X_test_selected)), "\n")
```

Testing the tuned models on the testing sets showed that the k-nearest neighbors regression model was the best ML model for our data set as it had the highest r-squared value and lowest RMSE value for the testing set. We also ran the tuned models with the training set to show how the hyperparameter tuning affected the scores. For the k-nearest neighbors model, this showed that the tuning eliminated the overfitting as the r-squared values and RMSE values for both the training and testing set became much closer to each other.

## 4. DISCUSSION

   We compared linear regression, ridge regression, lasso regression and k-nearest neighbors regression. The k-nearest neighbor regression algorithm had the best performance with an r-squared value of 0.564 and an RMSE value of 291.8 on the testing set. The ridge, lasso and linear regression algorithms had significantly worse performance with r-squared values of around 0.45 and RMSE values of around 256.2 for the testing set for each one. Based on these results, the k-nearest neighbor regression algorithm should be used for our predictive model. 
   
   The features in this dataset are not strong predictors of the outcome variable identified. While the r-squared value for the best-fitting algorithm is greater than 50%, it is by a small margin and does not correlate to significantly accurate predictions. Therefore, we cannot confidently believe that these features, using the algorithms we applied, will best predict outcome variabes for additional testing datasets.
   
   Because the p-values for both of our hypotheses was 0.00, which was notably lower than the alpha value of 0.05, we were able to reject the null hypotheses and confidently determine that a correlation existed between energy and popularity as well as danceability and popularity. With regards to our data, a significant difference means that the selected features do have an impact on song popularity; for the two hypotheses conducted, this means that energy and danceability were important features for the predictive model. A non-significant difference, on the other hand, poiints the fact that the null hypothesis cannot be rejected -- the selected features do not have an impact on song popularity.   
   
   In the future, better results may be revealed if additional streaming services are also analyzed to provide a more comprehensive picture of people's listening habits and preferences. Features not considered in this analysis, such as artist popularity/familiarity potentially also play a significant role in song popularity and could be considered for next time. It is also important to consider that, knowing the results of this project, it may be quite possible that the musical attributes of songs do not have any significant effect on their popularity, no matter how much additional data is received or analysis is conducted.
